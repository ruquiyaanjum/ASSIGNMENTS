3) App routing : * It is a process of mapping URLs to Python functions in a Flask application.
                 * App routing in Flask is like giving directions to different parts of your website .
                 * specify a URL parameter with a default value in Flask --  @app.route('/', defaults={'param': 'default'})
                 * we use app routes to tell Flask how to respond when someone visits a particular URL. 
                      For example, when a user goes to /home, Flask knows to show the home page, and
                      when they go to /about, it knows to show the about page.So, app routes help 
                      Flask know where to send users when they visit different parts of your site or app.

5) The function used in Flask for URL building is url_for().
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():

    return f'<a href="{url_for("about")}">Visit the about page</a>'

@app.route('/about')
def about():
    return 'This is the about page'

if __name__ == '__main__':
    app.run(host="0.0.0.0")
#In this code, url_for('about') generates the URL for the about function, which is then used to link to the about page.
