```python
##1)
#five functions:
import pandas as pd
#this function read the csv file.
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

#1) read first 5 records 
print(df.head())

#2) read last 5 records
print(df.tail())

#3) check the column data types
print(df.dtypes)

#4) describe -- numerical analysis based numeric types
print(df.describe())

#5) check the column names in file.
print(df.columns)




##2)

data ={
    "A":[10,20,30],
    "B":[40,50,60],
    "C":[70,80,90]}
df = pd.DataFrame(data)
#reindex which start with 1 and increment by 2
df.index=pd.RangeIndex(1,len(df)*2,2)
print(df)




##3)
data={
    "Values":[10,20,30,40,50]}
df = pd.DataFrame(data)
print(sum(df["Values"][:3]))



##4)
data={
    "Text":["sample text","yet another example","set the example"]}
df = pd.DataFrame(data)
df["Word_Count"]=df["Text"].apply(lambda x:len(str(x).split()))
print(df)




##5)
""" DataFrame.size(): multiply number of elements rows and columns.
    DataFrame.shape(): number of rows and columns.
    
"""
size=df.size
print(size)
shape=df.shape
print(shape)




##6)
#read_excel function to read excel file
df = pd.read_excel("Copy of LUSID Excel - Setting up your market data.xlsx")
print(df)




##7)
data = {
    "Email":["rukz@domain.com","anjum@domain.com","raja@domain.com"]}
df=pd.DataFrame(data)
df["Username"]=df["Email"].apply(lambda x:str(x).replace("@domain.com"," "))
print(df)




##8)
data = {
    "A":[3,8,6,2,9],
    "B":[5,2,9,3,1],
    "C":[1,7,4,5,2]}
df = pd.DataFrame(data)
print(df[(df["A"]>5) & (df["B"]<10)])



##9)

data={
    "Values":[10,20,30,40,50]}
df = pd.DataFrame(data)
print(df["Values"].mean())
print(df["Values"].median())
print(df["Values"].std())




##10)
data={
    "Sales":[100,120,110,90,95,105,115,130,125,110],
    "Date":pd.date_range(start='2023-01-01',periods=10)}
df=pd.DataFrame(data)
df["MovingAverage"]=df["Sales"].rolling(window=7,min_periods=1).mean()
print(df)




##11)
data={
    "Date":pd.date_range(start="2023-01-01",periods=5)
}
df=pd.DataFrame(data)
df["Weekday"]=df["Date"].dt.day_name()
print(df)



##12)
data={
    "Date":pd.date_range(start='2023-01-01',end='2023-01-31')
}
df=pd.DataFrame(data)
select_df=df[(df["Date"]>='2023-01-01') & (df["Date"]<='2023-01-31') ]
print(select_df)



##13)
"""the basic functions of pandas the first and foremost necessary library that needs to
be imported is pandas only
  import pandas as pd --- pd is alias refer to pandas function and classes
"""
```
